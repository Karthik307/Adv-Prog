```python
import matplotlib.pyplot as plt
a = np.arange(1,11) 
b = np.arange(11,21)
plt.xlabel("Array a Values")
plt.ylabel("Array b Values")
plt.title("a versus b")
plt.plot(a,b,color="red",linewidth=2)
plt.show()
```

```python
days = list(range(1,8))
temp = [34,36,40,41,42,34,44]
plt.xlabel("Days")
plt.ylabel("Temperatue")
plt.title("weather report")
plt.xlim([1,7])
plt.ylim([32,46])
plt.plot(days,temp,color="green",linewidth=2,linestyle="--")
plt.show()
```

```python
x=np.arange(10) #0to9
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,x,label="x values")
plt.plot(x,2*x,label="2*x values")
plt.plot(x,3*x,label="3*x values")
plt.plot(x,4*x,label="4*x values")
plt.legend()
plt.show()
```

## Bar plot

```python
sub=['C', 'C++', 'Java', 'Python', 'PHP'] 
students=[23,17,35,29,12]
preference = [90,85,90,95,80]
plt.bar(sub,students,color="green",width=0.5)
plt.xlabel("Languages")
plt.ylabel("Number of students")
plt.show()
```

```python
sub=['C', 'C++', 'Java', 'Python', 'PHP'] 
students=[23,17,35,29,12]
preference = [90,85,90,95,80]
plt.barh(sub,students,color="green")
plt.xlabel("Languages")
plt.ylabel("Number of students")
plt.show()
```

## 3.Subplots

```python
fig,a= plt.subplots(2,2) 
import numpy as np
x=np.arange(1,5) 
a[0][0].plot(x,x*x) 
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root') 
a[1][0].plot(x,np.exp(x)) 
a[1][0].set_title('exp') 
a[1][1].plot(x,np.log10(x)) 
a[1][1].set_title('log')
plt.show()
```

## 6.pie plot

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
# Show percentages for every pie. Specify radius to increase chart size
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=1.5)
plt.show()
```

```python
# Giving colors custom
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=1.5,colors = ['lightskyblue', 'red', 'blue', 'green', 'gold'] )
plt.show()
```

```python
#explode (seprate from pie chart)
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
#angle properties
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],startangle=45)
plt.show()
```
